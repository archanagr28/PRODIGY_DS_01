```python
total_population_df=df[df["Series Code"]=="SP.POP.TOTL"]
total_populated_sorted=total_population_df.sort_values(by="2021",ascending=False)
total_populated_sorted.head(10)

total_top_ten=total_populated_sorted.head(10)
print(total_top_ten["Country Code"])
total_populated_sorted.head(10)
```

```python
sns.barplot(x="2021",y="Country Code",hue="Country Code",data=total_top_ten,palette="coolwarm",legend=False)
plt.title("Top Ten Countries with Highest Population",fontsize=15)
plt.xlabel("Total population in the year 2021")
plt.ylabel("Countries")
plt.show()
```

```python
total_population_df1=df[df["Series Code"]=="SP.POP.TOTL"]
total_populated_sorted1=total_population_df1.sort_values(by="2021",ascending=True)
total_populated_sorted1.head(10)

total_bottom_ten=total_populated_sorted1.head(10)
print(total_bottom_ten["Country Code"])
total_populated_sorted1.head(10)
```

```python
sns.barplot(x="2021",y="Country Code",hue="Country Code",data=total_bottom_ten,palette="coolwarm",legend=False)
plt.xlabel("Bottom Ten Countries")
plt.ylabel("Country Code")
plt.show()
```
