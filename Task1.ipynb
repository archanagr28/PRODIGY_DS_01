```python
import pandas as pd
df=pd.read_csv("C:/Users/archa/Downloads/archive (3)/Population of India.csv")
df.head()

```

```python
import matplotlib.pyplot as plt
total_male=df["Male"].sum()
total_female=df["Female"].sum()
gender=["Male","Female"]
counts=[total_male,total_female]
plt.bar(gender,counts,color=['blue','pink'])
plt.title('Gender Distribution in Population')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Population[50]'], bins=10, color='green', edgecolor='black')
plt.title('Population Distribution by State')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```
